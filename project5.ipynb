## DAILY WEATHER DATA ANALYSIS

This Project analyzes daily weather data using python. The dataset is stored in a CSV file 
and analyzed using Pandas.Graphs are created using Matplotlib. 

##Import Libraries
We import the required Python libraries for data analysis and visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##  Load the Dataset 
In this step we load the weather dataset from a CSV file into a Pandas Dataframe.

```python
df = pd.read_csv("CSVFile Projects/daily_weather.csv")
df.head()
```

## Dataset Overview 
We check the structure of the dataset including column names, data types and summary ststistics.

```python
df.info()
df.describe()
```

## Dataset Overview
This dataset contains weather measurements recorded at 9 AM and 3AM.
The dataset includes variables such  as air temperature, humidity,wind speed etc 


```python
plt.plot(df["air_temp_9am"])
plt.title("Air Temperature at 9am")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```

```python
plt.plot( df["relative_humidity_9am"]) 

plt.title("Relative Humidity at 9AM")
plt.xlabel("Days")
plt.ylabel("Humidity")
plt.show()
```

```python
plt.plot(df["max_wind_speed_9am"])
plt.title("Wind Speed in 9AM")
plt.xlabel("Days")
plt.ylabel("Wind Speed")
plt.show()
```

```python
plt.plot(df["air_pressure_9am"])
plt.title("Air Pressure at 9AM")
plt.xlabel("Days")
plt.ylabel("Air Pressure")
plt.show()
```

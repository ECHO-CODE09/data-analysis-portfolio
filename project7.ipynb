 # Air Quality Index (AQI) Visualizer

This project analyzes AQI data across cities using Python, Pandas, and Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Dataset 
We loaded the Aqi Dataset using Pandas  

```python
df = pd.read_csv("CSVFile Projects/india_city_aqi_2015_2023.csv")
df.head()
```

## Data Cleaning 
Convert Data Column 
Removing Missing Value 

```python
df['date'] = pd.to_datetime(df['date'])
df = df.dropna()
df.info()
```

## Average AQI by city 
we group data by city and calculate mean AQI 

```python
city_group = df.groupby('city')['aqi'].mean().sort_values(ascending=False)
city_group.head()
```

## Plot AQI over time

This graph shows AQI changes  over time on different cities 


```python
plt.figure()

for city in df['city'].unique()[:5]:
    city_data = df[df['city'] == city]
    plt.plot(city_data['date'], city_data['aqi'], label = city)
    plt.xlabel('Date')
    plt.title("AQI changes over time")
    plt.ylabel('AQI')
    plt.legend()
    plt.xticks(rotation = 45)
    plt.show()
```

## Top 5 polluted cities 

We find cities with highest average AQI

```python

plt.figure()
plt.hist(df['aqi'], bins = 30)

plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.title('AQI Distribution')
plt.show()
```

## Conclusion 
- AQI varies across the cities
- Some cities have consistently high pollution 
- Trends help understand pollution patterns

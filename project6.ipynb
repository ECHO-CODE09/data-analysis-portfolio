#  Movie Ratings Explorer

This project explores a movie dataset using **Pandas** and **Matplotlib** to analyze movie ratings and genres.

The main objectives of this project are:

- Load and explore a movie ratings dataset
- Analyze average ratings by genre
- Visualize the distribution of ratings
- Identify the top 10 highest rated movies

This project demonstrates basic **data analysis and visualization using Python**.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##  Load the Dataset

We load the movie dataset using **Pandas**.  
The dataset contains information such as:

- Movie title
- Genre
- Rating
- Other metadata

```python
df = pd.read_csv("CSVFile Projects/imdb_1000.csv")

df.head()
```

##  Dataset Overview

Before performing analysis, we check:

- Number of rows and columns
- Column names
- Basic statistics

```python
df.info()
df.describe()
```

##  Average Rating by Genre

In this step we group movies by their **genre** and calculate the **average rating** for each genre.

This helps us understand which genre tends to receive higher ratings from viewers.

```python
genre_rating = df.groupby("genre")["star_rating"].mean()

genre_rating.sort_values(ascending=False)
```

##  Visualization: Average Rating by Genre

We create a **bar chart** to visualize the average ratings across different genres.

```python
genre_rating.sort_values().plot(kind='barh', figsize=(10,6))

plt.title("Average Movie Rating by Genre")
plt.xlabel("Average Rating")
plt.ylabel("Genre")

plt.show()
```

##  Distribution of Movie Ratings

A **histogram** helps us understand how movie ratings are distributed.

For example:
- Are most movies rated between 6–7?
- Are there many highly rated movies?

```python
plt.figure(figsize=(8,5))

plt.hist(df["star_rating"], bins=20)

plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Movies")

plt.show()
```

##  Top 10 Highest Rated Movies

Here we identify the **top 10 movies with the highest ratings** in the dataset.

```python
top_movies = df.sort_values("star_rating", ascending=False).head(10)

top_movies[["title","genre","star_rating"]]
```

##  Top 10 Movies Visualization

```python
plt.figure(figsize=(10,6))

plt.barh(top_movies["title"], top_movies["star_rating"])

plt.title("Top 10 Highest Rated Movies")
plt.xlabel("Rating")
plt.ylabel("Movie Title")

plt.gca().invert_yaxis()

plt.show()
```

##  Conclusion

From this analysis we observed:

- Certain genres tend to have higher average ratings.
- Movie ratings mostly fall within a specific range.
- A small number of movies achieve very high ratings.

This project demonstrates how **Pandas and Matplotlib** can be used for basic **data analysis and visualization**.
